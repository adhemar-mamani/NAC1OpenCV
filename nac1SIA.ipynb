```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
import numpy as np
import math

img = cv2.imread('circulos.png')
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cimg = cv2.cvtColor(gray_img,cv2.COLOR_GRAY2RGB)

circles = cv2.HoughCircles(gray_img, cv2.HOUGH_GRADIENT, dp=1.1, minDist=550, param1=200, param2=100, minRadius=70, maxRadius=90)

areas=[]

circulos=[]

if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0,:]:
        
        # desenha o contorno do circulo
        cv2.circle(cimg,(i[0],i[1]),i[2],(0,255,0),4)

        # desenha no centro do circulo
        cv2.circle(cimg,(i[0],i[1]),2,(0,0,255),5)
        
        area = math.pi * math.pow(i[2], 2)
       

        areas.append(area)
        circulos.append((i[0],i[1]))
        

for i in areas:
    print("Area do circulo: %.2f"%i)

plt.figure(figsize = (10,10))
plt.imshow(cimg, cmap="Greys_r", vmin=0, vmax=255)
plt.show()
```
